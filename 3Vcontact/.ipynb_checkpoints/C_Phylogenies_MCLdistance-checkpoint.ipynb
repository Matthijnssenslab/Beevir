```python
# This notebook processes the Runs through the phylogenies created from the protein clusters.


### Input files needed:
# - *raxml: raxml trees w/ created for the protein clusters.
# - c1.ntw: network file created by vcontact2.
# - vConTACT_proteins.csv: csv file containing the protein information of REFs.
# - pickle files with the distances get written out in the data folder as well.
```

```python
import ete3
import os
import itertools
import pandas as pd
import networkx as nx
import seaborn as sns
import numpy as np
import pickle
from progressbar import *
import scipy
from Bio.Phylo.TreeConstruction import DistanceCalculator
from Bio import AlignIO
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Get the branchlength distances
vogdists = {}
for i in os.listdir('data/'):
    if 'raxml' in str(i):
        vogname = i.split('.')[1]
        treename = 'data/' + i
        t = ete3.Tree(treename)
        R = t.get_midpoint_outgroup()
        t.set_outgroup(R)
        nodelist = []
        for node in t.traverse("postorder"):
            if 'NODE' in node.name:
                nodelist.append(node.name)
        iterres = itertools.combinations(nodelist, 2)
        distanceresult = []
        for j in iterres:
            res = t.get_distance(j[0], j[1])
            distanceresult.append([j[0], j[1], res])
        vogdists[vogname] = distanceresult
vogdists
```

```python
keepnodes = []
for i in vogdists:
    for j in vogdists[i]:
        if '_'.join(j[0].split('_')[0:8]) not in keepnodes:
            keepnodes.append('_'.join(j[0].split('_')[0:8]))
        if '_'.join(j[1].split('_')[0:8]) not in keepnodes:
            keepnodes.append('_'.join(j[1].split('_')[0:8]))
```

```python
#Pull vcontact network
#This takes quite some time....
#Vcontact network
vconnodes = []
vconedge = []
with open('data/c1.ntw') as f:
    for line in f:
        if line.strip().split()[0] not in vconnodes:
            vconnodes.append(line.strip().split()[0])
        if line.strip().split()[1] not in vconnodes:
            vconnodes.append(line.strip().split()[1])
        templis = [line.strip().split()[0], line.strip().split()[1], float(line.strip().split()[2])]
#        if templis not in vconedge:
        vconedge.append(templis)
```

```python
#Create graph
vcon_nxgraph = nx.Graph()
for i in vconnodes:
    vcon_nxgraph.add_node(i)
for i in vconedge:
    vcon_nxgraph.add_edge(i[0],i[1])
    vcon_nxgraph[i[0]][i[1]]['weight'] = i[2]
```

```python
nx.draw(vcon_nxgraph,with_labels=False,node_size=4, edge_color='lightgrey',pos=nx.spring_layout(vcon_nxgraph))
```

```python
overlapnodes = []
for i in vogdists:
    for j in vogdists[i]:
        if '_'.join(j[0].split('_')[0:8]) and '_'.join(j[1].split('_')[0:8]) in vconnodes:
            if '_'.join(j[0].split('_')[0:8]) not in overlapnodes:
                overlapnodes.append('_'.join(j[0].split('_')[0:8]))
            if '_'.join(j[1].split('_')[0:8]) not in overlapnodes:
                overlapnodes.append('_'.join(j[1].split('_')[0:8]))
overlapnodes
```

```python
nxgraphdists = []
widgets = ['collating branchlengths and bellman distances: ', Percentage(), ' ', Bar(marker='0',left='[',right=']'),' ', ETA(), ' ', FileTransferSpeed()]
pbar = ProgressBar(widgets=widgets, maxval=len(vogdists))
pbar.start()
counter = 0
for i in vogdists:
    counter += 1
    for j in vogdists[i]:
        try:
            quernam = '_'.join(j[0].split('_')[0:8])
            subjnam = '_'.join(j[1].split('_')[0:8])
            vconpath = nx.bellman_ford_path_length(vcon_nxgraph,quernam,subjnam, weight='weight')
            numedg = len(nx.shortest_path(vcon_nxgraph, source=quernam, target=subjnam, weight=None))
            #numedg = vcon_nxgraph.number_of_edges(quernam,subjnam)
            if numedg != 0:
                vconnum = vconpath / numedg
                nxgraphdists.append([i,'_'.join(j[0].split('_')[0:8]), '_'.join(j[1].split('_')[0:8]), j[2],vconpath,vconnum, numedg])
        except:
            pass
    pbar.update(counter)
pbar.finish()
```

```python
#Build DF:
quickpd = pd.DataFrame(nxgraphdists)
quickpd.columns = ['phylo','node1','node2','branchlength','dijkstralength','dijkstra/numedg','numedg']
quickpd['branchlog'] = np.log10(quickpd['branchlength'] + 1)
```

```python
sns.scatterplot(x="branchlength",y="dijkstra/numedg",data=quickpd, alpha=0.5)
scipy.stats.spearmanr(quickpd["branchlength"],quickpd["dijkstra/numedg"])
```

```python
########## ALL nodes (including refs for PC0000) ##########:
genomedic = {}
with open('data/vConTACT_proteins.csv') as f:
    for line in f:
        if 'NODE' not in line.strip().split(',')[0]:
            genomedic[line.strip().split(',')[0]] = line.strip().split(',')[1]
t = ete3.Tree('data/RAxML_bipartitions.PC_00000.nodes.refs.AL.TRIM.raxml')
R = t.get_midpoint_outgroup()
t.set_outgroup(R)
nodelistall = []
for node in t.traverse("postorder"):
    if len(node.name) != 0 and node.name not in nodelistall:
        nodelistall.append(node.name)
iterres = itertools.combinations(nodelistall, 2)
distanceresult = []
for j in iterres:
    res = t.get_distance(j[0], j[1])
    distanceresult.append([j[0], j[1], res])
distanceresult_genomename = []
for i in distanceresult:
    if 'NODE' in i[0] and 'NODE' in i[1]:
        distanceresult_genomename.append(i)
    elif 'NODE' in i[0] and 'NODE' not in i[1]:
        templis = [i[0], genomedic[i[1]], i[2]]
        distanceresult_genomename.append(templis)
    elif 'NODE' not in i[0] and 'NODE' not in i[1]:
        templis = [genomedic[i[0]], genomedic[i[1]], i[2]]
        distanceresult_genomename.append(templis)
    elif 'NODE' not in i[0] and 'NODE' in i[1]:
        templis = [genomedic[i[0]], i[1], i[2]]
        distanceresult_genomename.append(templis)
distanceresult_genomename
```

```python
nxgraphdists = []
widgets = ['Calculating dijkstras for REF & NODE:', Percentage(), ' ', Bar(marker='0',left='[',right=']'),' ', ETA(), ' ', FileTransferSpeed()]
pbar = ProgressBar(widgets=widgets, maxval=len(distanceresult_genomename))
pbar.start()
counter = 0
for j in distanceresult_genomename:
    counter += 1
    try:
        if 'NODE' in j[0]:
            quernam = '_'.join(j[0].split('_')[0:8])
        if 'NODE' not in j[0]:
            quernam = j[0].replace(' ', '~')
        if 'NODE' in j[1]:
            subjnam = '_'.join(j[1].split('_')[0:8])
        if 'NODE' not in j[1]:
            subjnam = j[1].replace(' ', '~')
        vconpath = nx.bellman_ford_path_length(vcon_nxgraph,quernam,subjnam, weight='weight')
        numedg = len(nx.shortest_path(vcon_nxgraph, source=quernam, target=subjnam, weight=None))
        #numedg = vcon_nxgraph.number_of_edges(quernam,subjnam)
        if numedg != 0:
            vconnum = vconpath / numedg
            if 'NODE' in j[0] and 'NODE' in j[1]:
                mixed = 'Nodes'
            if 'NODE' in j[0] and 'NODE' not in j[1] or 'NODE' not in j[0] and 'NODE' in j[1]:
                mixed = 'Mixed'
            if 'NODE' not in j[0] and 'NODE' not in j[1]:
                mixed = 'Ref'
            nxgraphdists.append([i,'_'.join(j[0].split('_')[0:8]), '_'.join(j[1].split('_')[0:8]), j[2],vconpath,vconnum, numedg, mixed])
    except:
        pass
    pbar.update(counter)
pbar.finish()
```

```python
quickpd = pd.DataFrame(nxgraphdists)
quickpd.columns = ['phylo','node1','node2','branchlength','dijkstralength','dijkstra/numedg','numedg', 'mixed']
quickpd['branchlog'] = np.log10(quickpd['branchlength'] + 1)
```

```python
sns.scatterplot(x="branchlength",y="dijkstra/numedg",data=quickpd,hue='mixed', alpha=0.5)
scipy.stats.spearmanr(quickpd["branchlength"],quickpd["dijkstra/numedg"])
```
