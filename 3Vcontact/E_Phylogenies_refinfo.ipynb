```python
from functools import reduce
import ete3
import dask.dataframe as dd
from dask.diagnostics import ProgressBar
import pandas as pd
import os
```

```python
#Grab tip labes from all trees
PC_tips = []
for i in os.listdir('data/'):
    if 'raxml' in i:
        treestr = 'data/' + str(i)
        t = ete3.Tree(treestr)
        for n in t.traverse():
            if len(n.name) > 1:
                PC_tips.append(n.name)
        PC_nodes = []
        PC_refs = []
        for i in PC_tips:
            if 'NODE' in i:
                PC_nodes.append(i)
            else:
                PC_refs.append(i)
```

```python
#Link refs with genomes (proteinlist from vcontact):
protdf = pd.read_csv("data/vConTACT_proteins_tilde.csv",sep=",", header=0)
protdf.columns = ["prot","Genome","keywords","clusters"]
protdf = protdf[protdf['prot'].isin(PC_refs)]
#Link the protdf with the genomes (to get tax information).
genomedf = pd.read_csv("data/genome_by_genome_overview.csv", sep=",", header=0)
prot_genome_df = pd.merge(protdf, genomedf, on="Genome", how="left")
prot_genome_df = prot_genome_df[['prot','Family']]
prot_genome_df = prot_genome_df.fillna('Unassigned')
```

```python
#PC0
#Subdivide per protein cluster for lists to use in R + fix syntax for listing.
PC_tips = []
t = ete3.Tree('data/RAxML_bipartitions.PC_00000.nodes.refs.AL.TRIM.raxml')
for n in t.traverse():
    if len(n.name) > 1:
        PC_tips.append(n.name)
PC_nodes = []
PC_refs = []
for i in PC_tips:
    if 'NODE' in i:
        PC_nodes.append(i)
    else:
        PC_refs.append(i)
len(prot_genome_df[prot_genome_df['prot'].isin(PC_refs)]) == len(PC_refs)
tempdf = prot_genome_df[prot_genome_df['prot'].isin(PC_refs)]
famdic = {}
for i in list(tempdf['Family'].unique()):
    famdic[i] = list(tempdf[tempdf['Family'] == i]['prot'])
for i in famdic:
    print(str(i)+ ' = c' + str(famdic[i]).replace('[', '(').replace(']',')'))
```

```python
print('Nodes = c' + str(PC_nodes).replace('[', '(').replace(']',')'))
```
